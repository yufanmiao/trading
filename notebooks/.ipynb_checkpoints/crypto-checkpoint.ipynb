```python
#simple linear regression as baseline
with open(data_path, 'r') as csvfile:
     reader = csv.reader(csvfile, delimiter=',')
     col_names = next(reader)
col_names.pop(1)
#dtypes = [str, int, float, float, float, float, float, float, str, str, str]
df_mkt_data = pd.read_csv(data_path, usecols = col_names, verbose=True)

print(df_mkt_data.dtypes)
#df_mkt_data.describe()
df_mkt_data['symbol'] = df_mkt_data['symbol'].astype('string')
df_mkt_data['buyerId'] = df_mkt_data['buyerId'].astype('string')
df_mkt_data['sellerId'] = df_mkt_data['sellerId'].astype('string')
df_mkt_data['type'] = df_mkt_data['type'].astype('string')
df_mkt_data['label'] = 0

df_mkt_data.loc[df_mkt_data['type'] == 'trade', 'label'] = 1

df_mkt_data['label'].value_counts()



```

```python
plot_cols = ['q_bid', 'p_bid', 'p_ask', 'q_ask', 'p_last', 'q_last', 'label']
plot_features = df_mkt_data[plot_cols]
plot_features.index = df_mkt_data['updateTime']
_=plot_features.plot(subplots=True)
```

```python
df_mkt_data.describe().transpose()
```

```python
features = df_mkt_data[plot_cols]
n = len(features)

train_dataset = features[0:int(n*0.7)]
val_dataset = features[int(n*0.7):int(n*0.9)]
test_dataset = features[int(n*0.9):]

n_features = features.shape[1]
```

```python
print(n_features)
features.head()
```

```python
#normalization
train_mean = train_dataset.mean()
train_std = train_dataset.std()

train_df = (train_dataset - train_mean)/train_std
val_df = (val_dataset - train_mean)/train_std
test_df = (test_dataset - train_mean)/train_std
```

```python
train_label = train_dataset['label']
train_data = train_df.iloc[:, : -1]
```

```python
train_data.shape
```

```python
print(train_label)
```

```python
#logistic to get the weights
model = keras.Sequential([
    layers.Dense(6, activation = 'relu', kernel_regularizer = regularizers.l1(l1=0.1),
                 #bias_regularizer = regularizers.l2(1e-4),
                 #activity_regularizer = regularizers.l2(1e-5)
                ),
    layers.Dense(1, activation = 'sigmoid')
    
]
)
```

```python
optimizer = tf.keras.optimizers.RMSprop(0.001)
model.compile(optimizer=optimizer, loss= 'mse', metrics=['mae', 'mse'])
```

```python
model.fit(train_data, train_label, epochs=100)
```

```python
model.summary()
```

```python
def ranges(nums):
    nums = sorted(set(nums))
    gaps = [[s, e] for s, e in zip(nums, nums[1:]) if s+1 < e]
    edges = iter(nums[:1] + sum(gaps, []) + nums[-1:])
    return list(zip(edges, edges))
```

```python
data_ranges = ranges(trade_row_idx)
windows = []
first_range = data_ranges[0]
if first_range[0] == first_range[1]:
    windows.append((0, first_range[0]))
else:
    windows.append((0, first_range[1]))

for i in range(1,len(data_ranges)):
    if data_ranges[i][0] == data_ranges[i][1]:
        windows.append((windows[i-1][1], data_ranges[i][0]))
    else:
        windows.append((windows[i-1][1], data_ranges[i][1]))
print(windows)



```

```python
split_rate = 0.8
train_range_size = int(np.ceil(len(windows)*split_rate))
x_train_range = windows[: train_range_size]
x_validation_range = windows[train_range_size:]
```
